## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

## Importing Datasets

```python
dataset = pd.read_csv('/content/drive/MyDrive/Machine Learning - Udemy/Part 3 - Classification/Section 14 - Logistic Regression/Python/breast_cancer.csv')
dataset.head(5)
```

### Checking for missing Values

```python
dataset.isnull()
```

```python
for col in dataset.columns:
  print((col , sum(dataset[col].isnull())))
```

```python
X = dataset.iloc[:,1:-1]
y = dataset.iloc[:,-1]
```

## Splitting the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

## Training the Dataset

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train, y_train)
```

## Predicting results for Test Set  

```python
y_pred = LR.predict(X_test)
y_pred
print(y_test)
```

```python
LR.score(X_train, y_train)
```

```python
LR.score(X_test, y_test)
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred)
print("{}".format(cm))
```

```python
cm_disp = ConfusionMatrixDisplay(confusion_matrix = cm)
cm_disp.plot()
plt.show()
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, y_pred, pos_label = 2)
```
