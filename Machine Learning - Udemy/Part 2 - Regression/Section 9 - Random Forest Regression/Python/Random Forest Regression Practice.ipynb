# Random Forest 

## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Dataset

```python
dataset = pd.read_csv('/content/drive/MyDrive/Machine Learning - Udemy/Part 2 - Regression/Section 8 - Decision Tree Regression/Python/Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

## Training the whole Dataset


```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators =10, random_state = 0)
rfr.fit(X,y)
```

## Predicting a new result

```python
rfr.predict([[6.5]])
```

```python
# only works for values within the range of the variable
rfr.predict([[12]]) 
```

## Visualizing Results

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y, color = 'red')
plt.plot(X, rfr.predict(X), color = 'blue')
plt.plot(X_grid, rfr.predict(X_grid), color = 'orange')
```
