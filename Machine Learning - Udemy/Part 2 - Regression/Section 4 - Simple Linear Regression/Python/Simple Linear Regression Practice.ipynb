# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
path = "/content/drive/MyDrive/Machine Learning - Udemy/Part 2 - Regression/Section 4 - Simple Linear Regression/Python/Salary_Data.csv"
dataset = pd.read_csv(path)
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,
                                                    random_state = 0)
```

# Training Simple Linear Regression on Training Set

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

# Predict for Test Dataset

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
# 12 is scalar
# [12] is 1D array / vector
# [[12]] is 2D array / matrix
# for predict method, input has to be a matrix
lr.predict([[12]])
```

# Visualizing the Training Set Results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, lr.predict(X_train), color = 'blue')
plt.title('Salary vs Experience - Training Set')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# Visualizing the Test Set results

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, lr.predict(X_train), color = 'blue') # As this plots a line
# The equation for line is same for Test and Train. But Check
#plt.plot(X_test, lr.predict(X_test), color = 'blue')
plt.title('Salary vs Experience - Test Set')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
