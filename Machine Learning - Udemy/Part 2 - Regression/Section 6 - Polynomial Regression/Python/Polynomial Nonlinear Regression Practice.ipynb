# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
dataset = pd.read_csv("/content/drive/MyDrive/Machine Learning - Udemy/Part 2 - Regression/Section 6 - Polynomial Regression/Python/Position_Salaries.csv")
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
print("X:\n{}".format(X))
print("y:\n{}".format(y))
```

# Linear Regression on the Whole Training Set

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
lr.score(X,y)
```

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, lr.predict(X), color = 'blue' )
```

# Polynomial Regression on the Whole Training Set

```python
plt.scatter(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
print("X_poly:\n{}".format(X_poly))
```

```python
# fitting Linear Regression on X_poly
plr = LinearRegression()
plr.fit(X_poly,y)
print("Accuracy:\t{}".format(plr.score(X_poly,y)))
print("Intercept:\n{}".format(plr.intercept_))
print("Coefficients:\n{}".format(plr.coef_))
```

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, plr.predict(X_poly), color = 'blue' )
```

# Visualization for smoother curve

```python
X_grid = np.arange(min(X), max(X), 0.1)
plt.scatter(X,y, color = 'red')
X_grid = X_grid.reshape((len(X_grid)),1)
plt.plot(X_grid, plr.predict(poly_reg.fit_transform(X_grid)), color = 'blue' )
```

# Predicting a a new Result

```python
lr.predict([[6.5]])
```

#### Polynomial

```python
plr.predict(poly_reg.fit_transform([[6.5]]))
```
